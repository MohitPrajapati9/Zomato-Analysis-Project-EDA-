### ***Filling Null values***

# ***Location Analysis***

```python
data.city.value_counts()
```

```python
City_count=data.city.value_counts().head(5)
Top_city=City_count.idxmax()
print("City with the most restaurant:",Top_city)

data.city.value_counts().head(5).plot(kind='bar',color='#E23744')
plt.title('Top 5 Cities by Restaurant Counts')
plt.xlabel("City")
plt.ylabel('Number of Restaurant')
plt.show()
```

# ***Rating Analysis***

```python
valid_ratings=data[data['aggregate_rating']>0]
Average_rating=round(valid_ratings['aggregate_rating'].mean(),2)
print("Average_rating:",Average_rating)

plt.figure(figsize=(12,6))
sns.histplot(data=data[data["aggregate_rating"]>0],x="aggregate_rating",color="#E23744",bins=20)
plt.axvline(Average_rating, color='black', linestyle='--', linewidth=2,label=Average_rating)
plt.xlabel("Aggregate Rating")
plt.ylabel("Count of Restaurants")
plt.title("Distribution of Restaurant Ratings")
plt.legend()
plt.show()
```

```python
valid_ratings=data[data['aggregate_rating']>0]
top5_cities = valid_ratings['city'].value_counts().head(5).index
average_rating = valid_ratings[valid_ratings['city'].isin(top5_cities)].groupby ('city')['aggregate_rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
average_rating.plot(kind='bar',color="#E23744")
plt.title("Average Restaurant Ratings in Top 5 Cities")
plt.xlabel("City")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```

# ***Cuisine Analysis***

```python
all_cuisines = data['cuisines'].dropna().str.split(', ')
cuisine_flat = all_cuisines.explode()
popular_cuisines = cuisine_flat.value_counts().head(10)

print("Top 10 Most Popular Cuisines:")
print(popular_cuisines)

plt.figure(figsize=(10, 5))
sns.barplot(x=popular_cuisines.values, y=popular_cuisines.index, palette="Reds_r")
plt.title("Top 10 Most Popular Cuisines", fontsize=14, color="#E23744")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.tight_layout()
plt.show()
```

# ***Price Range and Rating Analysis***

```python
avg_rating_by_price = df.groupby('price_range')['aggregate_rating'].mean()
plt.figure(figsize=(7, 5))
avg_rating_by_price.plot(kind='bar', color="#E23744")
plt.title("Average Rating by Price Range")
plt.xlabel("Price Range (1=Low, 4=High)")
plt.ylabel("Average Rating")
plt.show()
```

# ***Top Restaurant Chains***

```python
top_chains = df['name'].value_counts().head(10)
print("Top 10 Restaurant Chains by Number of Outlets:")
print(top_chains)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_chains.values, y=top_chains.index, palette="Reds_r")
plt.title("Top 10 Restaurant Chains by Outlet Count", fontsize=14)
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Chain")
plt.tight_layout()
plt.show()
```

```python
top_chain_names = top_chains.index
top_chain_data = df[data['name'].isin(top_chain_names)]

chain_avg_rating = top_chain_data.groupby('name')['aggregate_rating'].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=chain_avg_rating.values, y=chain_avg_rating.index, palette="Reds")
plt.title("Average Ratings of Top Restaurant Chains", fontsize=14)
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Chain")
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

#  ***Summary & Key Insights***

### **Summary** 
 This exploratory data analysis (EDA) aimed to uncover key trends and success factors in the restaurant industry using Zomato's restaurant dataset. We examined attributes such as location, ratings, pricing, cuisines and restaurant chains to better understand what influences customer experience and restaurant popularity.

### **Key Insights**

- **Location Matters** :
  - New Delhi, Bengaluru, and Mumbai have the highest concentration of restaurants.
  - These cities also show diverse rating distributions, indicating varying customer satisfaction.


- **Ratings Distribution** :
  - The average rating is approximately 3.65.
  - Majority of restaurants are rated between 3.0 and 4.5, suggesting moderate satisfaction.
  - Very few restaurants have extremely low or high ratings.


- **Cost vs Rating** :
  - Higher price ranges are slightly correlated with better ratings.
  - However, some low-cost restaurants also perform well, showing price isnâ€™t everything.
    

- **Cuisines & Popularity** :
  - North Indian, Chinese, and Fast Food are the most common cuisines.
  - Restaurants offering a diverse cuisine range tend to score better on ratings.
    

- **Top Chains** :
  - Dominoâ€™s, Pizza Hut, and McDonaldâ€™s dominate in terms of outlet numbers.
  - However, some smaller chains have better average ratings.
